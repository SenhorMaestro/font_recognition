```python
import pandas as pd
from PIL import Image, ImageFont, ImageDraw
import textwrap
import random
```

```python
df = pd.read_csv('IMDB Dataset.csv')
df['len'] = df['review'].apply(lambda x: len(x))
df = df[df['len']<700]
```

```python
font_filepaths = {0:"/home/nikinho/font_recognition/fonts/Aguante/Aguante-Regular.otf",
1:"/home/nikinho/font_recognition/fonts/Ambidexter/ambidexter_regular.otf",
2:"/home/nikinho/font_recognition/fonts/Aref Ruqaa Ink/ArefRuqaaInk-Bold.ttf",
3:"/home/nikinho/font_recognition/fonts/Aref Ruqaa Ink/ArefRuqaaInk-Regular.ttf",
4:"/home/nikinho/font_recognition/fonts/Better VCR/better-vcr-5.2.ttf",
5:"/home/nikinho/font_recognition/fonts/Brass Mono/BrassMono-Bold.ttf",
6:"/home/nikinho/font_recognition/fonts/Brass Mono/BrassMono-BoldItalic.ttf",
7:"/home/nikinho/font_recognition/fonts/Brass Mono/BrassMono-Italic.ttf",
8:"/home/nikinho/font_recognition/fonts/Brass Mono/BrassMono-Regular.ttf",
9:"/home/nikinho/font_recognition/fonts/Ganesha Type/GaneshaType-Regular.ttf",
10:"/home/nikinho/font_recognition/fonts/Ghastly Panic/GhastlyPanicCyr.otf",
11:"/home/nikinho/font_recognition/fonts/Realest/Realest-Extended.otf",
12:"/home/nikinho/font_recognition/fonts/Sans Collegiate One/AlumniSansCollegiateOne-Italic.ttf",
13:"/home/nikinho/font_recognition/fonts/Sans Collegiate One/AlumniSansCollegiateOne-Regular.ttf",
14:"/home/nikinho/font_recognition/fonts/Tana Uncial SP/TanaUncialSP.otf"
}
```

```python
font_colors = [(0,0,0), (128,128,128), (0,0,128), (128,0,128), 
               (0,128,128), (255,0,0), (128,0,0), (255,140,0)]
font_weights = [30, 30, 20, 4, 4, 4, 4, 4]
back_colors = [(255,255,255), (154,205,50), (0,0,255), (255,255,0), 
              (0,255,255), (0,128,0), (176,196,222), (255,255,240), 
              (240,255,255)]
back_weights = [40,10,10,6,6,6,6,10,6]
```

```python
folders = ['train', 'test']
folder_weights = [90, 10]
```

```python
def text_to_image(
text: str,
extra_space: int,
font_filepath: str,
font_size: int,
font_color: tuple[int, int, int],
back_color: tuple[int, int, int], 
font_align="left"):

   font = ImageFont.truetype(font_filepath, size=font_size)
   box = font.getsize_multiline(text)
   img = Image.new("RGB", (extra_space+box[0], extra_space+box[1]), back_color)
   draw = ImageDraw.Draw(img)
   draw_point = (extra_space/2, extra_space/2)
   draw.multiline_text(draw_point, text, font=font, fill=font_color, align=font_align)
   return img
```

```python
list_of_labels = []
```

```python
for i in range(10000):
    extra_space = random.randint(40, 80)
    line_width = random.randint(50, 75)
    folder = random.choices(folders, weights=folder_weights)[0]
    indent = random.choice(['\n', ''])
    text = df.iloc[i, 0]
    text = text.replace("<br /><br />", "\t")
    text = textwrap.fill(text, line_width, expand_tabs=True, tabsize=1000, 
                     subsequent_indent=indent) 
    font_size = random.randint(18, 28)
    font_class = random.randint(0, 14)
    font_filepath = font_filepaths[font_class]
    font_color = random.choices(font_colors, weights=font_weights)[0]
    back_color = random.choices(back_colors, weights=back_weights)[0]
    image = text_to_image(text, extra_space, font_filepath, font_size, font_color, back_color)
    image.save('/home/nikinho/font_recognition/images/%s/%s/image_%s.png' %(folder, font_class, i))
    list_of_labels.append((i, font_class, folder, image.size[0], image.size[1]))
```

```python
df_labels = pd.DataFrame(list_of_labels, 
        columns=['img_num','font_class','folder','width','height'])
```
